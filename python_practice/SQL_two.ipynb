```python
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
import pandas
from connect import info

engine = create_engine(info)

inspector = inspect(engine)
```

```python
Base = automap_base()
Base.prepare(engine, reflect=True)
Base.classes.keys()
```

```python
Film = Base.classes.film
```

```python
session = Session(engine)
```

```python
columns = inspector.get_columns('film')
for c in columns:
    print(c['name'])
```

```python
Film_finds = session.query(Film).filter(Film.title == 'Chamber Italian').all()
for Film_find in Film_finds:
    print("film_id: {0}, title: {1}".format(Film_find.film_id, Film_find.title))
```

```python
Film_finds = session.query(Film).filter(Film.title == 'Chamber Italian').all()
for Film_find in Film_finds:
    print("film_id: {0}, title: {1}".format(Film_find.film_id, Film_find.title ))   
```

```python
resut = session.query(Film).filter(Film.rental_rate < 2.00).\
order_by(Film.rental_rate.desc()).all()
```

```python
for r in resut:
    print(r.film_id, r.title, r.rental_rate)
```

```python
results = session.query(Film.film_id, Film.title, Film.rental_rate).filter(Film.rental_rate < 2.00).\
order_by(Film.rental_rate.desc()).all()
```

```python
df= pandas.DataFrame(results[:10], columns=['film_id','title', 'rental_rate'])
df.set_index('film_id', inplace=True)
df.head()
```
